```python
url = f"https://api.nasa.gov/neo/rest/v1/feed?api_key={API_KEY}"

response = requests.get(url)
data = response.json()
```

```python
print(response.status_code)
```

```python
## Loop through all dates and extract required asteroid information
all_asteroids = []

neos = data["near_earth_objects"]

for date in neos:
    for asteroid in neos[date]:

        asteroid_data = {
            "Asteroid ID": asteroid["id"],
            "Asteroid Name": asteroid["name"],
            "Min Diameter (km)": asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_min"],
            "Absolute Magnitude": asteroid["absolute_magnitude_h"],
            "Relative Velocity (km/s)": asteroid["close_approach_data"][0]["relative_velocity"]["kilometers_per_second"]
        }

        all_asteroids.append(asteroid_data)
```

```python
#Dataframe
import pandas as pd

df = pd.DataFrame(all_asteroids)

df.head()
```

```python
#Data cleaning
df = df.dropna()

df["Min Diameter (km)"] = df["Min Diameter (km)"].astype(float)
df["Relative Velocity (km/s)"] = df["Relative Velocity (km/s)"].astype(float)
```

```python
df.to_csv("asteroids.csv", index=False)
```
